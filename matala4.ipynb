```python
import requests
import re
from bs4 import BeautifulSoup
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt
import urllib.request, urllib.parse, urllib.error
```

```python
url = 'https://www.madlan.co.il/for-sale/הרצליה-ישראל'
html = requests.get(url)

headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/113.0.0.0 Safari/537.36'}
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
if response.status_code == 200:
    print('works')
else:
    print(f'The request failed: {response.status_code}')
```

```python
property_tags = soup.find_all(class_="css-zuyt5t e1sx3tzs1")
```

```python
list_for_df = []

for prop in property_tags:
    prop_dict = {} 
    city = "Herzliya"

    typs = prop.find("div", {"data-auto": "property-class"})
    typs_str = typs.get_text() if typs else "None"

    rooms = prop.find('div', {'data-auto': 'property-rooms'})
    rooms_str = rooms.get_text() if rooms else "None"

    area = prop.find('div', {'data-auto': 'property-size'})
    area_str = area.get_text() if area else "None"

    adress = prop.find('div', {'data-auto': 'property-address'})
    adress_str = adress.get_text(strip=True) if adress else "None"
    list_elements = adress_str.split(",")

    if list_elements[0] != adress_str:
        street_and_num = list_elements[0]
        city_area = list_elements[1]
    elif adress_str == "None":
        street_and_num = "None"
        city_area = "None"
    else:
        # In this code, I refer to the extreme case noted above
        city_area = list_elements[0]
        street_and_num = "None"

    street_and_num_list = street_and_num.split()
    street = ''
    num = ''
    for i in street_and_num_list:
        if i.isnumeric():
            num = num + i
        else:
            street = street + ' ' + i

    num = int(num) if num != '' else "None"
    street = street if street != '' else "None"

    price = prop.find("div", {"data-auto": "property-price"}) 
    price_str = price.get_text() if price else "None"
    if price_str != "None":
        price_int = re.findall(r'\d+(?:,\d+)*', price_str)
        price_int = int(price_int[0].replace(',', ''))
    else:
        price_int = "None"

    list_for_df.append({
        "City": city,
        "Type": typs_str,
        "room_number": rooms_str,
        "Area": area_str,
        "Street": street,
        "street_number": num,
        "city_area": city_area,
        "Price": price_int
    })

df = pd.DataFrame(list_for_df)
```
